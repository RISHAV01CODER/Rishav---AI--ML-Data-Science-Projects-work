```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import seaborn as sns
```

```python
test_df = pd.read_csv("test.csv")
train_df = pd.read_csv("C:/Users/souga/OneDrive/Documents/train.csv")
train_df.head()
```

```python
X = train_df.drop(columns=['Price','ID','date'])
Y = train_df['Price']
X
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.1)
```

```python
reg = LinearRegression()
reg.fit(X_train,Y_train)
```

```python
test_prediction = reg.predict(X_test)
actual_dataprice = Y_test
```

```python
print(mean_squared_error(actual_dataprice, test_prediction))
```

```python
id = test_df["ID"]
test_df_priceprediction = test_df.drop(columns=["date","ID"]) #the test dataset which we'll feed to the 
predicted_price = reg.predict(test_df_priceprediction)
```

```python
submission_df = pd.DataFrame()
submission_df["ID"] = id
submission_df["Price"] = predicted_price
submission_df.set_index("ID", inplace = True)
```
